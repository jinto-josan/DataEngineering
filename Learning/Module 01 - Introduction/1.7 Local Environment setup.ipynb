## Local Environment setup

In this lesson you will be able to have your local environment set up. Despite the fact that you'll be using Databricks in your learning journey and you'll use notebooks; sometimes it's helpful to have your local machine ready to keep learning, practicing and coding in Python and Apache Spark.

**⚠️** **IMPORTANT**: these instructions apply for MAC OS but might be tried in a Linux distribution too.

## Learning Objectives
By the end of this lesson, you should be able to
* Have installed Python, Java Development Kit (JDK), Apache Spark, and PyCharm CE IDE in your local environment

🕒 **Estimated Time For Completion:** 25 minutes

### Prerequisites MAC OS

Make sure you’ve installed **[homebrew](https://brew.sh/)** package manager

You can use a different Shell such as **Zsh** ([Oh My Zsh!](https://ohmyz.sh/)) or others where bash profile referred in this guide will belong to the corresponding file (in default bash your bash profile is located on `**~/.bash_profile**` if you use Zsh it’s located on `**~/.zshrc**`)


### Before Starting

For the local machine set up we’re going to use **Spark-3.0.2 with Hadoop2.7**, each **Apache Spark** distribution works with a **specific Python and Java version** so it’s important to review the compatibility matrix for the version that you’ll use and install the required versions for Java and Python:

- [Python Compatibility Matrix for Spark Distribution](https://community.cloudera.com/t5/Community-Articles/Spark-Python-Supportability-Matrix/ta-p/379144)
- [Java Compatibility Matrix for Spark Distribution](https://community.cloudera.com/t5/Community-Articles/Spark-and-Java-versions-Supportability-Matrix/ta-p/383669)

For the version that we’ll work on we'll need **Python v3.9** and **JDK v1.8**.


### Installing Python 3.9.6
1. Open your terminal and execute 
> `brew install python@3.9.6`

2. In your terminal execute
> `python3` or `python` if you have set up that as an alias on your system 
<br/>
  You should have entered to the Python console
  <br><img src="assets/python.png">
  <br><p>To exit the REPL, simply type <code>quit()</code> and press the enter key.</p>

3. Close the Python console and install the **`Py4J`** package using **`pip`**
> `python -m pip install "py4j"`

### Installing JDK Java Version 8

1. Download the JDK of Java Version 8 from the following [**link**](https://www.oracle.com/in/java/technologies/javase/javase8u211-later-archive-downloads.html) 
<br/>**IMPORTANT NOTES:** **A)** You have to register with Oracle using an emailof your choice. In addition, if you are working with a **Mac with Apple M1 Pro Chipset** download the file `jdk-8u391-macosx-aarch64.dmg` else `jdk-8u391-macosx-x64.dmg`. **Note** If you get a `400` error while accessing the above link, please try clearing the website data, including cookies and try again. Also, if `jdk-8u391` is not avaiable at the above link or has reached the End of Life, please try downloading and using the latest `jdk-8xxxx` version. It should work alright for the setup below. **B)** If you don't want to register with Oracle, or have trouble logging in, you can try installing the JDK via `brew install --cask corretto@8` to install the Amazon Corretto version of the JDK.
1. Execute the DMG (if you haven't used brew before) file to install the downloaded JDK
1. If you need to work with different Java versions, install **[jenv](https://www.jenv.be/)**
- Open your terminal and execute 
> `brew install jenv`

- Open your bash profile and add the following environment variables
> `export PATH="$HOME/.jenv/bin:$PATH"`<br>
> `eval "$(jenv init -)"`<br>

- In your terminal, locate your installed Java Virtual Machines executing `sudo find / -name Java` (Most of the time you’ll find them on `/Library/Java/JavaVirtualMachines`)

- Add the virtual machines to **jenv** executing the following command: 
> `jenv add /Library/Java/JavaVirtualMachines/jdk-1.8.jdk/Contents/Home`

- List your virtual machines in **jenv** using the following command: 
> `jenv versions`
- Select the Java Version for the shell using the following command
> `jenv global <the_name_of_your_jvm>`

- Don’t forget to source your bash profile
> `source .bash_profile (Default Bash)`<br>
> `source .zshrc (For Zsh)`

4. Finally execute `java -version` and make sure that you are using the `v1.8.0_xxx`<br/>
<br><img src="assets/jdk18.png"><br>



### Installing Apache Spark
1. Download [**spark-3.0.2-bin-hadoop2.7.tgz**](https://archive.apache.org/dist/spark/spark-3.0.2/spark-3.0.2-bin-hadoop2.7.tgz)
1. Open your terminal, go the the folder where you downloaded file and decompress it using the following command:
> `tar -xvzf spark-3.0.2-bin-hadoop2.7.tgz`
>> **Attention**: If you downloaded the file via a browser, it may have been unzipped automatically.
>> In this case untar it only: `tar -xvf spark-3.0.2-bin-hadoop2.7.tar`

1. Run the command `pwd` and copy the full path
> `For example "/data-tools/spark-3.0.2-bin-hadoop2.7”`
1. Open your bash profile and add the following environment variables
> `export SPARK_HOME="/data-tools/spark-3.0.2-bin-hadoop7"`<br>
> `export PYSPARK_PYTHON=python3.9.6`<br>
> `export PATH="$HOME/.jenv/bin:$PATH:$SPARK_HOME/bin"`<br>
1. Source your bash profile
1. Finally, in your terminal execute `pyspark` and you should be able to open the PySpark Console as shown below:
<br><br><img src="assets/pyspark.png">

1. Configure **PyCharm** following these [instructions](https://github.com/data-derp/exercise-vanilla-spark/tree/main?tab=readme-ov-file#repo-set-up-in-pycharm)
1. Restart **PyCharm** to have all set up
