## OLAP vs OLTP

Up until now, you have learnt about data engineering and the common roles that you find in a data team. In this lesson you'll get a better understanding of whether the datastore is **OLTP** or **OLAP** in nature.\

## Learning Objectives
By the end of this lesson, you should be able to:
* Know the differences between OLTP and OLAP.
* Get and understanding that when we are dealing with a data project or product we are mainly referring to OLAP-style workloads.

**OLTP vs OLAP**

|   | **OLTP - Online Transactional Processing** | **OLAP - Online Analytical Processing** | 
|---| ---  | ---  |
| _Purpose_ |support daily transactions|report and analyze data|
| _Design_ |application-oriented|subject-oriented|
| _Data_ |up-to-date, operational|consolidated, historical|
| _Size_ |Gigabytes|Terabytes, Exabytes, Pettabytes|
| _Queries_ |simple transactions and frequent updates|complex, aggregated queries and limited updates|
| _Users_ |thousands|hundreds|
| _Tasks Examples_ |Update the price of a book|Get the books with best profit margin|

**Database vs Data Warehouse**

|   | **Database** | **Data Warehouse** | 
|---| ---  | ---  |
| _Processing Method_ |OLTP|OLAP|
| _Frequent operations_ |DELETE, INSERT, UPDATE quickly in transactions|SELECT and analyze massive volumes of data|
| _Data Structure_ |Highly normalized data|Denormalized|
| _Data Timeline_ |Current, real-time data|Historical|
| _Data Analysis_ |Slow and painful due to the large number of table joins|Analysis is fast and easy due to the small number of table joins|

**Data Warehouse vs Data Lake**

|   | **Data Warehouse** | **Data Lake** | 
|---| ---  | ---  |
| _Data_ |structured, processed|raw, structured, non-structured, semi-structured|
| _Processing_ |schema-on-write|schema-on-read|
| _Storage_ |expensive for large data volumes|designed for low-cost storage|
| _Agility_ |less agile, fixed configuration|configure and reconfigure as needed|
| _Security_ |Mature|Maturing|
| _Users_ |Business Professionals|Data Scientists, ML Engineers|


- Organizations use Data Warehouses and Data Lakes to **store**, **manage** and **analyze data**

- **Data Lake:** A data lake is a highly scalable storage system that holds **structured** and **unstructured** data in its original form and format. A data lake does not require planning or prior knowledge of the data analysis - it assumes that analysis will happen later, on-demand.

- **Data Warehouse:** Traditional data warehouses use a process called **_Extract-Transform-Load (ETL)_**. Data is meticulously mapped from the original data sources to tables in the Data Warehouse, and undergoes transformations to achieve a structured format, to enable reporting and data analysis.


**⚠️** **IMPORTANT:** When people/customers talk about developing **Data & AI**, **Analytics**, **Data Science**, **Data Engineering**, **Machine Learning**, they’re most likely referring to **OLAP-style workloads**.
