## How Data Engineering Works

Watch the following [video](https://www.youtube.com/watch?v=qWru-b6m030) to get an overall introduction of what data engineering is.

💡 **REMEMBER:**

- **Data engineering** is the development, implementation, and maintenance of systems and processes that take in raw data and produce high-quality, consistent information that supports downstream use cases, such as analysis and machine learning.

- **Data Warehouse:** is a data repository where you have structured and historical data for a business domain, this data is consumed to analyze.

- **Data Lake:** is a data repository where you have structured and non-structured data (videos, images, text, etc...)

- **Cloud Data Warehouse:** some cloud vendors have evolved data warehouses offering highly scalable computing and storage capabilities and storage in the cloud. Some examples are: [_Amazon Redshift_](https://aws.amazon.com/redshift/), [_Google Big Query_](https://cloud.google.com/bigquery), [_Azure Synapse_](https://azure.microsoft.com/en-us/products/synapse-analytics)

- **ETL:** stands for _Extract - Transform - Load_, it's a pattern to ingest data commonly used to load data into Data Warehouses.

- **ELT:** stands for _Extract - Load - Transform_, it's a modern pattern to ingest data and commonly used to load data into Data Lakes.

- **Data Pipeline:** is the combination of architecture, systems, patterns, and processes that moves and transforms data to the desired destinations (Data Warehouse, Data Lake or others).

- **Big Data:** mainly refers to the volume or data size (_Terabytes, Petabytes, Exabyte, Zetabytes_) that needs to be processed and analyzed. Thanks to modern technologies and highly scalable computing power now we are able to process and analyze it. This has become an integral part of data engineering.

- **Batch and Streaming**: are ingestion processes used to move the data to the required destinations (Data Warehouse, Data Lake, and so on). Batch usually ingests the data in a daily frequency. Streaming ingests the data in real-time, once it's generated at the source.
