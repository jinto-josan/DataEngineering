## Setting up Apache Kafka in Local (Bonus)




**Note** This is a bonus exercise. Please do it only if you are running on schedule !

Let us set up Apache kafka in our local machines now to get some practice. You can see the instructions at [this link](https://kafka.apache.org/quickstart) for the same. Please note the following before you start doing this set up :

1. Make sure that you have java 8+ installed in your system before proceeding.
2. There are two ways of setting up this up as given in the instructions. a) using kraft and b) using Zookeeper. We suggest you follow the Zookeeper based installation as kraft might need you to install Docker / Colima additionally.
3. You just need to follow the instructions till `Step 5` in the above link. `Steps 6 & 7` are pretty advanced and can be ignored. Once `step 5` is completed, you can move to `step 8` to wind up the cluster and exercise.
